```python
import pandas as pd
import plotly.express as px
```

```python
car_data = pd.read_csv("/Users/Usuario/Documents/Proyectos/Car_Sales/vehicles_us.csv")
print(car_data.info())
```

```python
print('CANTIDAD DE VALORES AUSENTES POR COLUMNA:')
print(car_data.isna().sum())
print()
print('CANTIDAD DE FILAS DUPLICADAS:')
print(car_data.duplicated().sum())
```

```python
car_data['model_year'] = car_data['model_year'].fillna('0')
car_data['model_year'] = car_data['model_year'].astype('int')
car_data['cylinders'] = car_data['cylinders'].fillna(0)
car_data['cylinders'] = car_data['cylinders'].astype('int')
car_data['odometer'] = car_data['odometer'].fillna(0)
car_data['odometer'] = car_data['odometer'].astype('int')
car_data['paint_color'] = car_data['paint_color'].fillna("Not available")
car_data['is_4wd'] = car_data['is_4wd'].fillna("No")
car_data['is_4wd'] = car_data['is_4wd'].replace(1, "Yes")
print(car_data.sample(5))
```

### Se realizó análisis inicial del dataset, detectando lo siguiente:

1. Se encontraron datos ausentes en 5 de las 13 columnas contenidas en el archivo.
2. En las columnas correspondientes al año del vehiculo 'model_year', número del vehículo 'cylinders' y valor del odómetro 'odometer', se sustituyeron los valores ausentes por cero "0", ya que las columnas contienen valores numéricos de tipo entero.
3. Las columnas 'model_year' y 'odometer' fueron convertidas en tipo Entero, ya que la información que suministran no contiene y no requiere expresarse en decimales.
4. Para los valores ausentes de la columna 'paint_color', se llenaron los valores ausentes con el texto "Not available", para mantener su información en el tipo String.
5. En el caso de la columna que informa si el vehículo es o no de tracción 4FW 'is_4wd', se detectó que la información que suministra es tipo booleana pero estaba tipificada en número entero. Se presume que todas las filas con el número 1 indicaban que el vehículo si es 4wd por lo que se reemplazó por el texto "Yes", mientras que el resto (valores ausentes) se rellenaron con el texto "No".
